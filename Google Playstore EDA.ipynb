### 4.1 Data Preprocessing

```python
playstore_df.sort_values(by = 'Last Updated', inplace = True)
playstore_df.head()
```

```python
playstore_df.info()
```

```python
playstore_df.describe()
```

```python
#Finding duplicate rows based on App Name
duplicate = playstore_df[playstore_df.duplicated('App')]
duplicate
```

```python
#Removing duplicates
playstore_df.drop_duplicates(subset = ['App'], keep = 'last',inplace=True)
playstore_df.info()
```

```python
playstore_df['Installs'].unique()
```

```python
#Standardizing Reviews and No of Installations
playstore_df['Reviews'] = playstore_df['Reviews']/1000
playstore_df['No of Installs'] = playstore_df['Installs'].str.replace('[+,,]', '').astype(float)/1000000
playstore_df.head()
```

```python
#Standardizing App Size
playstore_df['Size in M'] = playstore_df['Size'].str[:-1].replace('Varies',np.nan,regex=True).astype(float).where(playstore_df['Size'].str[-1] != 'k',playstore_df['Size'].str[:-1].replace('Varies',np.nan,regex=True).astype(float)/1000)
playstore_df.info()
```

```python
playstore_df.describe()
```

```python
playstore_df.groupby(['Category'])['Rating'].mean()
```

```python
playstore_df.groupby(['Category'])['Rating'].apply(lambda x: x.isnull().sum())
```

```python
#Filling missing values for Rating by Category wise average
playstore_df['Rating'] = playstore_df.groupby(['Category'], sort=False)['Rating'].apply(lambda x: x.fillna(x.mean()))
```

```python
playstore_df.groupby(['Category'])['Rating'].apply(lambda x: x.isnull().sum())
```

```python
#Creating Revenue column (in Mn)
playstore_df['Revenue']=playstore_df['Price']*playstore_df['No of Installs']
```

```python
#Creating a Categorical Variable for Rating
playstore_df['Rating Category'] = pd.cut(playstore_df['Rating'],
                         [0,1,2,3,4,5],
                         labels=['< 1','1-2','2-3','3-4','4-5'], include_lowest=True)
playstore_df.head()
```

# <a id=section50103></a>
### 5.1.3 What is the contribution of the top 3/10/15/20 categories by count of apps?

```python
from collections import Counter
Category_count = Counter(playstore_df['Category'].dropna().tolist()).most_common(20)
Category_idx = [Category[0] for Category in Category_count]
Category_val = [Category[1] for Category in Category_count]
Categories = pd.concat([playstore_df.loc[playstore_df['Category']=='FAMILY'], playstore_df.loc[playstore_df['Category']=='GAME'], 
                       playstore_df.loc[playstore_df['Category']=='TOOLS']]).reset_index(drop=True)
print('There consists {} Apps from the top 3 categories out of the {} apps that is {}%'
      .format(Categories.shape[0], playstore_df.shape[0],np.round(Categories.shape[0]/playstore_df.shape[0]*100,decimals=2)))
Category_count_10 = Counter(playstore_df['Category'].dropna().tolist()).most_common(10)
Top10Cat = pd.Series( Category[0] for Category in Category_count_10 )
Category_count_15 = Counter(playstore_df['Category'].dropna().tolist()).most_common(15)
Top15Cat = pd.Series( Category[0] for Category in Category_count_15 )
Category_count_20 = Counter(playstore_df['Category'].dropna().tolist()).most_common(20)
Top20Cat = pd.Series( Category[0] for Category in Category_count_20 )
ps_10 = playstore_df[playstore_df['Category'].isin(Top10Cat)]
print('There consists {} Apps from the top 10 categories out of the {} apps that is {}%'
      .format(ps_10.shape[0], playstore_df.shape[0],np.round(ps_10.shape[0]/playstore_df.shape[0]*100,decimals=2)))
ps_15 = playstore_df[playstore_df['Category'].isin(Top15Cat)]
print('There consists {} Apps from the top 15 categories out of the {} apps that is {}%'
      .format(ps_15.shape[0], playstore_df.shape[0],np.round(ps_15.shape[0]/playstore_df.shape[0]*100,decimals=2)))
ps_20 = playstore_df[playstore_df['Category'].isin(Top20Cat)]
print('There consists {} Apps from the top 20 categories out of the {} apps that is {}%'
      .format(ps_20.shape[0], playstore_df.shape[0],np.round(ps_20.shape[0]/playstore_df.shape[0]*100,decimals=2)))
```
